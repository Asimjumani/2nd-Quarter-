```python
size_200 = (200,200)

for f in os.listdir("."):
    if f.endswith(".jpg"):
        i = Image.open(f)
        fn,fext = os.path.splitext(f)
        
        i.thumbnail(size_200)
        i.save("/size_200{}_200{}".format(fn,fext))
```
